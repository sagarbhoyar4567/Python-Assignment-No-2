```python
import statistics
Values=[]

while True:
    i=int(input('Provide your Values'))
    if i!=0:
        Values.append(i)
    if i==0:
        if   len(Values)==0:
             print('Error : You have entered 0 in the first attempt')
             break
        elif len(Values)!=0:
             Avg=statistics.mean(Values)
             print('Average of the list ',Avg)
             Values.clear()
             break
```

```python
string=input('Provide the input String')
string2=''
for i in string[::-1]:
    string2=string2+i
#print(string2)    
if string==string2:
    print(f'Provided Input string {string} is palindrome')
```

```python
import math
value=input('Provide the Binary Value :')
total=0
j=0
for i in value[::-1]:
    total=total+math.pow(2,j)*int(i)
    j+=1
print(f'provided binary No is {value} and converted Decimal value is {total}')
    
```
